```python
df = new_data[new_data['Order ID'].duplicated(keep = False)]
df["order together"] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df = df[['Order ID' , 'order together']].drop_duplicates()
df.head(5)
```

What are those products that sold together most

```python
from itertools import combinations
from collections import Counter
count = Counter()
for row in df['order together']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key , value in count.most_common(10):
    print(key,value)

# As we can see in output the product that sold most in pair of two is iphone and lighting charging cable

```

what product sold the most and why do you think it sold the most

```python
product_group = new_data.groupby('Product')
quantity_ordered  = new_data.groupby('Product')['Quantity Ordered'].sum()

products = [product for product , df in product_group]

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation= 'vertical',size = 8)
plt.xlabel('Products')
plt.ylabel('Quantity ordered')
plt.show()

# we can see the AAA Batteries (4- pack) sold the most
```

```python
prices = new_data.groupby('Product').mean()['Price Each']
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products,quantity_ordered,color = 'g')
ax2.plot(products,prices,'b-')
ax1.set_xlabel('Product name')
ax1.set_ylabel('Quantity Ordered',color = 'g')
ax2.set_ylabel('Price $',color = 'b')
ax1.set_xticklabels(products, rotation= 'vertical',size = 8)
plt.show()

# so the price of AAA batteries are very low comparison with other products thic can be the reason of sold the most AAA batteries
```

```python

def count_iphone_orders(products):
   
    product_list = products.split(',')
  
    return "iPhone" in product_list

new_data['Is_iPhone_Order'] = new_data['Product'].apply(count_iphone_orders)

total_orders = len(new_data)
iphone_orders = new_data['Is_iPhone_Order'].sum()
probability = iphone_orders / total_orders


probability_percentage = probability * 100

print(f"The probability that the next person will order an iPhone is {probability_percentage:.2f}%")
```
