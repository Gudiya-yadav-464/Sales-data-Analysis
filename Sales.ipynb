```python
dataframe = []
for files in os.listdir(path_dir):
    if files.endswith(".csv"):
        file_path = os.path.join(path_dir,files)
        df = pd.read_csv(file_path)
        dataframe.append(df)
```

```python
import re 
cities = []
city_pattern = r',\s(.*?),'
for index , row in data.iterrows():
    purchase_address = row["Purchase Address"]
    match = re.search(city_pattern,purchase_address)
    if match:
        city = match.group(1)
        cities.append(city)
    else:
        cities.append(None)

data['city'] = cities
print(data)
```
